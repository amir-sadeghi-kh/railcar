```python
data =EOCC_data.drop(['CAR NUMBER','key','LOCATION','RESPONSE CODE','WHY MADE','PART APPLIED DATE','UPDATED PART REPAIR DATE','UPDATED PART REPAIR DATE','BUILT DATE'],axis =1)
```

```python
data['CAR_CLASS'] = pd.factorize(data.CAR_CLASS)[0]
data['EQUIP_FEAT_DESC'] = pd.factorize(data.EQUIP_FEAT_DESC)[0]
data['car_length'] = pd.factorize(data.car_length)[0]
data['car_tonnage'] = pd.factorize(data.car_tonnage)[0]
data['BUILDER_NAME'] = pd.factorize(data.BUILDER_NAME)[0]
```

```python
df_new = pd.DataFrame(columns=['CAR_CLASS',
                               'EQUIP_FEAT_DESC',
                              'CONDITION CODE',
                              'BUILDER_NAME',
                              'EOCC_POCKET',
                              'weeks_in_service',
                              'Mileage_since_last _failure',
                               'car_age',
                               'Avg Mileage New Replace',
                               'Count of New Replacement',
                               'Avg Mileage Recond Replace',
                               'Count of Recond Replacement',
                               'Car life to date mileage',
                               'Loading_Count',
                               'TARGET'
                              ])
```

```python
df_new['CAR_CLASS'] = data['CAR_CLASS']
df_new['EQUIP_FEAT_DESC'] = data['EQUIP_FEAT_DESC']
df_new['CONDITION CODE'] = data['CONDITION CODE']
df_new['BUILDER_NAME'] = data['BUILDER_NAME']
df_new['EOCC_POCKET'] = data['EOCC_POCKET']
df_new['weeks_in_service'] = data['weeks_in_service']
df_new['Mileage_since_last _failure'] = data['Mileage_since_last _failure']
df_new['car_age'] = data['car_age']
df_new['Avg Mileage New Replace'] = data['Avg Mileage New Replace']
df_new['Count of New Replacement'] = data['Count of New Replacement']
df_new['Avg Mileage Recond Replace'] = data['Avg Mileage Recond Replace']
df_new['Count of Recond Replacement'] = data['Count of Recond Replacement']
df_new['Car life to date mileage'] = data['Car life to date mileage']
df_new['Loading_Count'] = data['Loading_Count']
df_new['TARGET'] = data['TARGET']
```

```python
y = df_new['TARGET']
x = df_new.drop(['TARGET'],axis =1)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix

x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=0)
```

```python
Hidden_layer_sizes = [50, 70, 90, 100, 130, 150, 200, 300]
Learning_rate_init = [0.01, 0.001, 0.0001]
Max_iter = [100, 200, 300, 500, 1000]
Learning_rate = ['constant', 'invscaling', 'adaptive']
epsilon = [1e-7, 1.5e-7, 1e-8, 1.5e-8, 1e-9]
```

```python
for h in Hidden_layer_sizes:
    for l in Learning_rate_init:
        for m in Max_iter:
            for lr in Learning_rate:
                for e in epsilon:
                    clf = MLPClassifier(solver='adam', hidden_layer_sizes = h, learning_rate_init = l, max_iter = m, learning_rate = lr, random_state=0,epsilon =e)
                    clf.fit(x_train,y_train) 
                    y_predict_test = clf.predict(x_test)
                    if confusion_matrix(y_test, y_predict_test)[1,1] > 40:
                        print(confusion_matrix(y_test, y_predict_test))
                        print('{0}-{1}-{2}-{3}-{4}'.format(h,l,m,lr,e))
                        a1 = confusion_matrix(y_test, y_predict_test)[0,0]/(confusion_matrix(y_test, y_predict_test)[0,0] + confusion_matrix(y_test, y_predict_test)[0,1])
                        b1 = confusion_matrix(y_test, y_predict_test)[1,1]/(confusion_matrix(y_test, y_predict_test)[1,1] + confusion_matrix(y_test, y_predict_test)[1,0])
                        print ('accuracy is {0}'.format((a1+b1)/2))
```
